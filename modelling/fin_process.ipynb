```python
# news_id_df
fin_df = pd.concat([df_naver, df_daum], axis=0)
```

```python
fin_df
```

```python
# 인덱스 초기화
fin_df.reset_index(drop=True, inplace=True)
```

```python
fin_df
```

```python
# fin_df 저장
import pickle 

with open('fin_df', 'wb') as f:  # 경로주의
    pickle.dump(fin_df, f)
```

```python
# 형태소 분석
from konlpy.tag import Mecab

mecab = Mecab()

tokenized_df = [mecab.morphs(content) for content in fin_df['content']]
```

```python
from gensim.models import Doc2Vec
from gensim.models.doc2vec import TaggedDocument


# TaggedDocument 객체 생성
tagged_df = [TaggedDocument(doc, [i]) for i, doc in enumerate(tokenized_df)]

# Doc2Vec 모델 생성
model = Doc2Vec(vector_size=300, min_count=2, epochs=40)

# 모델 학습
model.build_vocab(tagged_df)
model.train(tagged_df, total_examples=model.corpus_count, epochs=model.epochs)
```

```python
# 학습된 모델 저장
model.save('doc2vec_fin_model.model')
```

```python
tmp_list = [model.docvecs.most_similar(id, topn=20) for id in tqdm(fin_df.index)]
news_id = dict(zip(range(len(tmp_list)), tmp_list))
```

```python
# 저장
import pickle

with open('fin_df_recommand_list.pkl', 'wb') as f:
    pickle.dump(tmp_list, f)
with open('news_id.pkl', 'wb') as f:
    pickle.dump(news_id, f)
```

```python
recommand_df = pd.concat([fin_df, pd.Series(tmp_list, name='recommand')], axis=1)
recommand_df.head(1)
```

```python
# 유사도 인덱스값만 추출 
recommand_df.recommand = recommand_df.recommand.apply(lambda x: [_[0] for _ in x])
```

```python
# 정렬
recommand_df.recommand = recommand_df.recommand.apply(lambda x: sorted([_[0] for _ in x]))
```

```python
# 저장
with open('recommand_df.pkl', 'wb') as f:
    pickle.dump(recommand_df, f)
```
