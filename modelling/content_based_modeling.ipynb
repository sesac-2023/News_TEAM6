# naver news 

```python
# naver news
import pandas as pd

df = []

for n in range(5):
    df.append(pd.DataFrame(pd.read_pickle(f'./dataset/naver_news/total_news_data_0{n}.pkl')))
```

```python
df_naver = pd.concat(df, axis=0)
```

```python
df_naver.head(2)
```

```python
df_naver.reset_index(drop=True, inplace=True)
```

```python
# naver news 전처리
import pandas as pd
import pickle
import re

ch_df = df_naver[df_naver[1].apply(lambda x: False if x.find('@')==-1 else True)][1].str.split('@').str[0].apply(lambda x: ' '.join(re.split(r'[(, )]', x)[:-1]))
df_naver.loc[ch_df.index,1] = ch_df
df_naver[9] = df_naver[9].apply(lambda x: ' '.join([_ for _ in x if '\t' not in _ and '\n' not in _]))
df_naver[12] = '세계'
df_naver = df_naver[[12,5,8,2,0,1,3,4,9,11,7]]
df_columns = ['cat1_name', 'cat2_name', 'platform_name', 'title', 'press', 'writer', 'date_upload', 'date_fix', 'content', 'sticker', 'url']
df_naver.columns = df_columns

# 결측값은 모두 None으로 변경
df_naver.loc[df_naver[df_naver['date_fix']=='null'].index, 'date_fix'] = None
df_naver.loc[df_naver[df_naver['writer']=='null'].index, 'writer'] = None
```

```python
df_naver.head(2)
```

```python
import pandas as pd

# naver news 본문만 클렌징된 데이터 
naver_content = pd.read_excel('./dataset/naver_news/naver_content_final.xlsx') # 경로
```

```python
naver_content
```

```python
naver_content['content']
```

```python
# 클렌징된 본문으로 교체
df_naver['content'] = naver_content['content']
```

```python
df_naver[df_naver['content'].isna()].head(2)
```

```python
# 본문 결측값 기사 제목으로 대체
df_naver.loc[df_naver[df_naver['content'].isna()].index, 'content'] = df_naver[df_naver['content'].isna()].title
df_naver[df_naver['content'].isna()]
```

```python
from konlpy.tag import Mecab

mecab = Mecab()

tokenized_df = [mecab.morphs(df) for df in df_naver['content']]
```

```python
from gensim.models import Doc2Vec
from gensim.models.doc2vec import TaggedDocument


# TaggedDocument 객체 생성
tagged_df = [TaggedDocument(doc, [i]) for i, doc in enumerate(tokenized_df)]

# Doc2Vec 모델 생성
model = Doc2Vec(vector_size=300, min_count=2, epochs=40)

# 저장된 모델 불러오기 
model = Doc2Vec.load('doc2vec_model.model')

# 모델 학습
model.build_vocab(tagged_df)
model.train(tagged_df, total_examples=model.corpus_count, epochs=model.epochs)
```

```python
a = 276
b = 5

news = []
for i in model.docvecs.most_similar(a, topn=b):
    print(i)
    #print(i[0])
    #print(df_daum.iloc[i[0]])
    #print(pd.DataFrame(df_daum.iloc[i[0]]))
    news.append(i[0])
#print(news)
```

```python
df_naver.iloc[276]
```

```python
pd.DataFrame(df_naver.iloc[[news[0], news[1], news[2], news[3], news[4]]]).T
```

# naver + daum 

```python
# naver news, daum news concat
df_news = pd.concat([df_naver, df_daum], axis=0)
```

```python
# 인덱스 초기화
df_news.reset_index(drop=True, inplace=True)
```

```python
# df_news 저장
import pickle 

with open('../dataset/df_news.pkl', 'wb') as f:  # 경로주의
    pickle.dump(df_news, f)
```

```python
# df_news 불러오기 
with open('../dataset/df_news.pkl', 'rb') as f:  # 경로주의
    df_news = pickle.load(f)
```

```python
# 형태소 분석
from konlpy.tag import Mecab

mecab = Mecab()

tokenized_df = [mecab.morphs(content) for content in df_news['content']]
```

```python
from gensim.models import Doc2Vec
from gensim.models.doc2vec import TaggedDocument


# TaggedDocument 객체 생성
tagged_df = [TaggedDocument(doc, [i]) for i, doc in enumerate(tokenized_df)]

# Doc2Vec 모델 생성
model = Doc2Vec(vector_size=300, min_count=2, epochs=40)

# 모델 학습
model.build_vocab(tagged_df)
model.train(tagged_df, total_examples=model.corpus_count, epochs=model.epochs)
```

```python
# 학습된 모델 저장
model.save('doc2vec_news_model.model')
```
